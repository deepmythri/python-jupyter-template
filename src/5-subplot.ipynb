### Exercise 5: Subplot Layout

```python
# TASK: Create a 2x2 subplot layout.
# Plot a line chart in the first subplot, a bar chart in the second, a scatter plot in the third, and a pie chart in the fourth.
import matplotlib.pyplot as plt
import numpy as np

# Create a figure and 2x2 subplot layout
fig, axs = plt.subplots(2, 2, figsize=(10, 10))

# Subplot 1: Line chart
x = np.arange(0, 11, 1)
y = x**2
axs[0, 0].plot(x, y, marker='o', color='blue')
axs[0, 0].set_title("Line Chart")
axs[0, 0].set_xlabel("x values")
axs[0, 0].set_ylabel("y values")
axs[0, 0].grid(True)

# Subplot 2: Bar chart
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]
colors = ['red', 'blue', 'green', 'orange']
axs[0, 1].bar(categories, values, color=colors)
axs[0, 1].set_title("Bar Chart")
axs[0, 1].set_xlabel("Categories")
axs[0, 1].set_ylabel("Values")

# Subplot 3: Scatter plot
x_scatter = [1, 2, 3, 4, 5]
y_scatter = [2, 4, 6, 8, 10]
axs[1, 0].scatter(x_scatter, y_scatter, color='purple')
for i in range(len(x_scatter)):
    axs[1, 0].annotate(f"({x_scatter[i]}, {y_scatter[i]})", 
                       (x_scatter[i], y_scatter[i]), 
                       textcoords="offset points", 
                       xytext=(5, 5), 
                       ha='center')
axs[1, 0].set_title("Scatter Plot")
axs[1, 0].set_xlabel("x values")
axs[1, 0].set_ylabel("y values")
axs[1, 0].grid(True)

# Subplot 4: Pie chart
labels = ['Python', 'Java', 'C++', 'JavaScript']
sizes = [40, 25, 20, 15]
axs[1, 1].pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
axs[1, 1].set_title("Pie Chart")

# Adjust layout for better visualization
plt.tight_layout()
plt.show()
```
